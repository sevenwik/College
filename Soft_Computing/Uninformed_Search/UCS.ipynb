```python
import pandas as pd
from graphviz import Graph

def UCS(edgeList,initialNode, goalNode):
    Closed = []
    Open1 = []
    Open1.append(initialNode)
    dot = Graph(comment='UCS Algo')
    pathCost = []
    edgeCost = 0
    Path = ''
    Cost = 0
    traversed = []
    Open = [[initialNode,0]]
    #print(pathCost)
    count = 0
    currentNode = initialNode
    currentPathCost = initialNode
    while(Open1 or currentNode == ''):
        Open.pop(0)
        Open1.pop(0)
        if(currentNode == goalNode):
            break
        Closed.append(currentNode)

        uniqueFrom = []
        childFrom = edgeList[(edgeList["from"]==currentNode)]
        for i in childFrom['to']:
            if i not in Closed:
                uniqueFrom.append(i)

        uniqueTo = []
        childTo = edgeList[(edgeList['to']==currentNode)]
        for i in childTo['from']:
            if i not in Closed:
                uniqueTo.append(i)

        childNodePath = []
        #childNodeCost = []

        for i in uniqueFrom:
                Open1.append(i)
                #print(i)
                Open.append([i,list(edgeList[(edgeList['from']==currentNode) & (edgeList['to']==i)]['cost'])[0]+edgeCost])
                childNodePath = [currentPathCost+i,list(edgeList[(edgeList['from']==currentNode) & (edgeList['to']==i)]['cost'])[0]+edgeCost,list(edgeList[(edgeList['from']==currentNode) & (edgeList['to']==i)]['cost'])[0]+edgeCost]
                pathCost.append(childNodePath)
                if currentNode+i not in traversed:
                    traversed.append(currentNode+i)
                    dot.edge(currentNode,i,constraint='true',label = str(" ")+str(list(edgeList[(edgeList['from']==currentNode) & (edgeList['to']==i)]['cost'])[0]+edgeCost))
                childNodePath = []

        for i in uniqueTo:
                Open1.append(i)
                Open.append([i,list(edgeList[(edgeList['to']==currentNode) & (edgeList['from']==i)]['cost'])[0]+edgeCost])
                childNodePath = [currentPathCost+i,list(edgeList[(edgeList['to']==currentNode) & (edgeList['from']==i)]['cost'])[0]+edgeCost,list(edgeList[(edgeList['to']==currentNode) & (edgeList['from']==i)]['cost'])[0]+edgeCost]
                pathCost.append(childNodePath)
                if currentNode+i not in traversed:
                    traversed.append(currentNode+i)
                    dot.edge(currentNode,i,constraint='true',label = str(" ")+str(list(edgeList[(edgeList['to']==currentNode) & (edgeList['from']==i)]['cost'])[0]+edgeCost))
                childNodePath = []

        #print(pathCost)
        #print(Open)
        pathCost.sort(key=lambda x:x[2])
        Open.sort(key=lambda x:x[1])
        #print("After Sort")
        #print(pathCost)
        #print(Open)
        currentNode = Open[0][0]
        currentPathCost = pathCost[0][0]
        Path = currentPathCost
        Cost = pathCost[0][2]
        edgeCost = pathCost[0][1]
        #print(edgeCost)
        pathCost.pop(0)
        #print("=================================")   
    #print(traversed)
    for i in Path:
        dot.node(i,color="red")
    dot.node(goalNode,color="green") 
    #dot.edges(traversed)
    dot.attr(label=r"\n\n UCS Algorithm Graph")
    dot.render('output/UCS.gv', view=True)  
    'output/UCS.gv.pdf'
    print("============COST============")
    print(Cost)
    return Path



edgeList = pd.read_csv('edge_list.csv')
#straigtLine = pd.read_csv('straight_line_distance.csv')
#print(straigtLine['node'])

A_Algo_Path = UCS(edgeList,"A","M")
goalNode = 'M'
print("============PATH============")
for i in range(0,len(A_Algo_Path)-1):
    print(A_Algo_Path[i],'--> ',end="")
print(goalNode)
```
