### Question-2
      Histogram for the sugar content of the cereals in each shelf.


```python
topShelf = data[(data['shelf']=='Top')]
middleShelf = data[(data['shelf'] == 'Middle')]
bottomShelf = data[(data['shelf'] == 'Bottom')]

fig, axis = plt.subplots(1,3)
#fig.suptitle('Histogram plots for sugar content in cereals')

axis[0].hist(topShelf['sugars'])
axis[0].set_title("Top Shelf")

axis[1].hist(middleShelf['sugars'])
axis[1].set_title("Middle Shelf")

axis[2].hist(bottomShelf['sugars'])
axis[2].set_title("Bottom Shelf")

topShelf['sugars'].mean(), middleShelf['sugars'].mean(), bottomShelf['sugars'].mean() 
```

### Question-4
      5-number summary,mean and standard deviation for the variable Fiber

```python

Fiber = data['fiber']
quartiles = percentile(Fiber, [25, 50, 75])
data_min, data_max = Fiber.min(), Fiber.max()
print("5 Number Summary: ")
print('  Min: ', data_min)
print('  Q1: ', quartiles[0])
print('  Median: ', quartiles[1])
print('  Q3: ', quartiles[2])
print('  Max: ', data_max)

print("Mean: %.2f"%Fiber.mean())
print("Standard Deviation: %.2f"%numpy.std(Fiber))
```

### Question-5
      Scatterplot of the 1 varibles(Calories vs Carbohydrates)

```python
x = data['carbo']
y = data['calories']

plt.title("Scatterplot - Calories VS Carbohydrates")
plt.xlabel("Carbohydrates")
plt.ylabel("Calories")
plt.scatter(x,y)
plt.show()
corr = data['carbo'].corr(data['calories'])
print(corr)
```

Based on the correalation we can say that both the variables change in the same direction, because they have a positive correlation. So, we can say that they have Weak Positive correlation since the range is within 0.1-0.3.
